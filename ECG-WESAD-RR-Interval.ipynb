```python
import numpy as np
import biosppy.signals

# Load data dari file
data = np.loadtxt('C:/Users/Faiz/Downloads/Kuliah/UTS/S2_respiban.txt')

# Mengambil data ecg
ecg_data = data[:, 2]

# Menghitung rpeaks dan rr intervals
rpeaks = biosppy.signals.ecg.ecg(signal=ecg_data, sampling_rate=700, show=False)['rpeaks']
rr_intervals = np.diff(rpeaks)/700

# Menampilkan hasil
np.set_printoptions(threshold=np.inf)
print("Detected S2 R peaks:", rpeaks)
print("Detected S2 RR intervals:", rr_intervals)
```
